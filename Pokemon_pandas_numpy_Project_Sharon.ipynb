## Parte 2: Análisis con Numpy  

11. Convierte las columnas de estadísticas (`HP`, `Attack`, `Defense`, `Sp. Atk`, `Sp. Def`, `Speed`) en un array de Numpy.  

```python
import numpy as np
df_Pokemon_Numpy = df_Pokemon.copy()
df_Pokemon_Numpy = df_Pokemon_Numpy.to_numpy()
array_stats = df_Pokemon_Numpy[:, 5:11]
print(array_stats)
```

12. Calcula la media, mediana y desviación estándar de cada estadística usando Numpy.  

```python
columnas_de_estadisticas = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
for i, columna in enumerate(columnas_de_estadisticas):
    media = np.mean(array_stats[:, i])
    mediana = np.median(array_stats[:, i])
    desviacion_estandar = np.std(array_stats[:, i])
    print(f'Estadísticas para {columna}:')
    print(f'  Media: {media}')
    print(f'  Mediana: {mediana}')
    print(f'  Desviación Estándar: {desviacion_estandar}')
    print("-" * 20)
```

13. Normaliza las estadísticas dividiendo cada valor por el máximo de su columna.

```python
# Calcular el máximo valor
máximo_stats = np.max(array_stats, axis=0)

# Dividiendo cada valor por el máximo de su columna.
Resultado_normalizado = array_stats / máximo_stats

print(Resultado_normalizado)
```

14. Encuentra los índices de los 5 Pokémon con mayor defensa (`Defense`).  

```python
indices_top_5 = np.argsort(array_stats[:, 2])[-5:]
print(f'Los índices de los 5 Pokémon con mayor defensa son: {indices_top_5}')
```

15. Calcula la correlación entre `Attack` y `Defense`.

```python
attack = df_Pokemon["Attack"].to_numpy()
defense = df_Pokemon["Defense"].to_numpy()
corre = np.corrcoef(attack,defense)[0,1]
print(f"La corrlación entre Defensa y Ataque es: {round(corre,4)}")
```

16. Usa Numpy para encontrar cuántos Pokémon tienen una suma de (`HP`, `Attack` y `Defense`) superior a 300.

```python
# (HP, Attack y Defense)
hp_attack_defense = array_stats[:, :3]

suma = np.sum(hp_attack_defense, axis=1)

count_greater_than_300 = np.sum(suma > 300)

print(f"El numero de Pokémons con una suma mayor a 300 de HP, Attack y Defense son: {count_greater_than_300}")
```
